```python
import pandas as pd
import os
from tqdm.notebook import trange, tqdm

from nba_api.stats.static import teams
from nba_api.stats.endpoints import leaguegamefinder
```

```python
nba_teams = teams.get_teams()
```

```python
team_name_append = []
team_id_append = []

for i in nba_teams:
    team_name_append.append(i["full_name"])
    team_id_append.append(i["id"])

team_table = pd.DataFrame({"team_name":team_name_append, "team_id":team_id_append})
team_table.to_csv("data/team_ids.csv", index=False)
team_table
```

```python
for index, row in tqdm(team_table.iterrows()):
    team_id = row["team_id"]
    team_name = row["team_name"].lower().replace(" ", "_")
    team_name = team_name + "_" + str(team_id)
    
    save_filepath = f"data/{team_name}"
    os.mkdir(save_filepath)

    gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=team_id)
    games = gamefinder.get_data_frames()[0]

    output_filepath = os.path.join(save_filepath, team_name + ".csv")

    games.to_csv(output_filepath, index=False)
```
