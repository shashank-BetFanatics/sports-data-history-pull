```python
# When using SR data, please cite us and provide a link and/or a mention.
# https://www.pro-football-reference.com/boxscores/game-scores.htm

import pandas as pd
import requests
from bs4 import BeautifulSoup
from tqdm.notebook import trange, tqdm
```

```python
# Pull the games data organized by box score

r = requests.get("https://www.pro-football-reference.com/boxscores/game-scores.htm")
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
# Create a list of URLs to scrape

# Test version where i manually downloaded the webpage
# with open("data/nfl_data/games_by_scores.html") as fp:
#     soup = BeautifulSoup(fp, 'html.parser')

url_list = []

for i in soup.find_all("table")[0].find_all(class_="center"):
    if i.a:
        url = "https://www.pro-football-reference.com" + i.a["href"]
        url_list.append(url)
```

```python
box_scores_list = []

for i in tqdm(url_list):
    box_scores_list.append(pd.read_html(i)[0])


box_scores_output = pd.concat(box_scores_list)
box_scores_output
```

```python
box_scores_output.to_csv("data/nfl_data/nfl_box_scores.csv", index=False)
```
